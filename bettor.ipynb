Importar les llibreries que utilitzarem. En aquest cas, random

https://docs.python.org/2/library/random.html#module-random

Aquesta llibreria conté les eines necessàries per manipular i generar nombres aleatoris.

Funcions més generals:

1. Obtenir un nombre aleatori N tal que a <= N <= b:
```python
random.randint(a,b) 
```
2. Escollir aleatòriament un membre d'una llista seq:
```python
random.choice(seq)
```
3. Generar un nombre aleatori entre 0.0 i 1.0
```python
random.random()
```
4. Generar un nombre aleatori uniformement distribuït entre a i b
```python
random.uniform(a,b)
```
5. Generar un nombre aleatori seguint una gaussiana $G(\mu,\sigma)$
```python
random.gauss(mu, sigma)
```

Definim una funció *rollDice* que genera un nombre N entre 1 i 100.

Si N <= 50 o N és igual a 100, el jugador perd. En cas contrari, el jugador guanya. En cada cas, un missatge s'escriu en pantalla i es retorna True o False en cas de guanyar o perdre respectivament.


````
def rollDice():
    roll = random.randint(1,100)

    if roll == 100:
        print roll,'roll was 100, you lose. What are the odds?! Play again!'
        return False
    elif roll <= 50:
        print roll,'roll was 1-50, you lose.'
        return False
    elif 100 > roll >= 50:
        print roll,'roll was 51-99, you win! *pretty lights flash* (play more!)'
        return True
````

Es defineix una funció *simple_bettor* que rep tres paràmetres:


*   **funds**: Quantitat de diners inicial
*   **initial_wager**: Quantitat de diners per apostar
*   **wager_count**: Nombre de vegades que es farà l'aposta

Aquesta funció crida **wager_count** vegades a *rollDice* apostant **initial_wager** i en cas de guanyar (perdre) la quantitat corresponent s'afegeix (es redueix) a **funds**. 



````
def simple_bettor(funds,initial_wager,wager_count):
    value = funds
    wager = initial_wager

    currentWager = 0

    while currentWager < wager_count:
        if rollDice():
            value += wager
        else:
            value -= wager

        currentWager += 1
        print 'Funds:', value
````
