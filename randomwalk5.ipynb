```
##
##  This is a random walk problem in two dimensions coded in Python
##  E. Fernandez, Dec. 2015
##
##  Fill numpy arrays with random numbers 1,0, -1
##  These arrays are produced for each random walk
##  Then the coordinates x and y and the distances to the origin are
##  calculated for each step.
##  The x and y coordinates are written to disk for each random walk, while the distances
##  are summed to then calculate the average distance over several random walks.
##  The averages are then writen to disk.
##  The files with x and y are called xy[number].txt where [number] stands
##  for random walk 'number'.
##  The average distances are writen in file AvgDist

import numpy as np
import random
import matplotlib.pyplot as plt

# ask for number of random walks

print (35 * '-')
print ("   THIS IS THE 2-d RANDOM WALK PROGRAM")
print (35 * '-')


## Get input: number of random walks (to take average of distances) and   
## numnber of steps for each random walk

nber_rdwks = int ( input('Enter number of Random Walks (maximum 20) : ') )
                
if 1<= nber_rdwks <= 20:     #check that numbers is between bounds
    print ("We will generate", nber_rdwks, " Random Walks")
else :
    print (nber_rdwks, " is out of range, try again")
    stop

nber_steps = int ( input('Enter number of steps in each Random Walk (maximum 200000) : ') )

if 1<= nber_steps <= 20000:  #check that numbers is between bounds
    print ("We will generate", nber_steps, " steps in each Random Walk")
else :
    print (nber_steps, " is out of range, try again")
    stop

# Define array of average distances and fill it with zeros.
# The averages are over number of random walks
avg_dist = np.zeros(nber_rdwks)
avg_distance = 0.

for k in range (0, nber_rdwks):
    i=k+1
    x_array = np.random.randint(-1,2,size=nber_steps)
    y_array = np.random.randint(-1,2,size=nber_steps)
    
    x_coord = np.zeros((nber_steps,),dtype=np.int)
    y_coord = np.zeros((nber_steps,),dtype=np.int)
    distance = np.zeros(nber_steps, dtype=float)
    
    print (" this is Random Walk number ", i)
    #print (" these are the first 10 elemnets of x_array ", x_array[0:10])
    #print (" these are the first 10 elemnets of y_array ", y_array[0:10])
    
    # Open file for xy coordinates
    file1 = 'xy' + str(i) + '.txt'
    f1 = open(file1,'w')
    
    print ("We start with loop in number of generated points = ", nber_rdwks, " for random walk ", i)
    newcoordx = 0
    newcoordy = 0
   
   #plt.figure(i)
    for index in range(nber_steps):
        
        x_coord[index] = newcoordx + x_array[index]
        y_coord[index] = newcoordy + y_array[index]
        
        #print ( "index= ", index, newcoordx, newcoordy, x_array[index], y_array[index], x_coord[index], y_coord[index])
        
        #l = plt.axhline(y=newcoordy, xmin=newcoordx, xmax=x_coord[index])
        #l = plt.axvline(x=x_coord[index], ymin=newcoordy, ymax=y_coord[index])
        
        distance[index] = (x_coord[index]**2+y_coord[index]**2)**0.5
        
        newcoordx = x_coord[index]
        newcoordy = y_coord[index]
        
        s1 = "{0:d} {1:d} {2:d} {3:d} {4:.3f} \n ".format(x_array[index],y_array[index],x_coord[index],y_coord[index],distance[index])
    
        #print (s1)
        # Write x and y coordinates to disk
        f1.write(s1)
    # make a quick plot
    plt.plot(x_coord,y_coord)

    f1.close()
    avg_dist[k] = np.average(distance)
    plt.show()

avg_distance = np.average(avg_dist)
print ("Grand average distance = ", avg_distance)
   
```
