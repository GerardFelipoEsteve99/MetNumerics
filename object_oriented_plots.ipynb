```
# Plotting object_oriented style

import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(1)  # a new figure window
ax = fig.add_subplot(1, 1, 1)  # specify (nrows, ncols, axnum)

x = np.linspace(0, 10, 1000)
y = np.sin(x)
ax.plot(x, y)

plt.show()

fig = plt.figure(2)
ax = fig.add_subplot(1, 1, 1)  # specify (nrows, ncols, axnum)

y2 = np.cos(x)
ax.plot(x, y)
ax.plot(x, y2)

plt.show()

fig = plt.figure(3)
ax = fig.add_subplot(1, 1, 1)  # specify (nrows, ncols, axnum)

ax.plot(x, y)
ax.plot(x, y2)
ax.set_ylim(-1.5, 2.0)

ax.legend(['sine', 'cosine'])

plt.show()

fig = plt.figure(4)
ax = fig.add_subplot(1, 1, 1)  # specify (nrows, ncols, axnum)

ax.plot(x, y)
ax.plot(x, y2)
ax.set_ylim(-1.5, 2.0)

ax.set_xlabel("$x$")
ax.set_ylabel("$\sin(x)$")
ax.set_title("I like $\pi$")

plt.show()

fig = plt.figure(5)

for i in range(6):
    ax = fig.add_subplot(2, 3, i + 1)
    ax.set_title("Plot #%i" % i)

fig.subplots_adjust(wspace=0.3, hspace=0.3)

x = np.linspace(0, 10, 1000)
for i in range(6):
    fig.axes[i].plot(x, np.sin(i * x))

plt.show()

# create just a single figure and axes

#fig = plt.figure(6)

fig, ax = plt.subplots()
ax.plot(x, np.sin(x))

plt.show()

fig, ax = plt.subplots(2, 3)  # 2x3 grid

for i in range(2):
    for j in range(3):
        ax[i, j].plot(x, np.sin((3 * i + j) * x))

plt.show()

fig, ax = plt.subplots(2, 3, sharex=True, sharey=True)  # 2x3 grid

for i in range(2):
    for j in range(3):
        ax[i, j].plot(x, np.sin((3 * i + j) * x))

plt.show()

fig = plt.figure(figsize=(8, 8))
gs = plt.GridSpec(3, 3)
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, :2])
ax3 = fig.add_subplot(gs[1:, 2])
ax4 = fig.add_subplot(gs[2, 0])
ax5 = fig.add_subplot(gs[2, 1])

plt.show()

fig = plt.figure(figsize=(8, 8))
gs = plt.GridSpec(3, 3)
ax1 = fig.add_subplot(gs[0, :])
ax2 = fig.add_subplot(gs[1, :2])
ax3 = fig.add_subplot(gs[1:, 2])
ax4 = fig.add_subplot(gs[2, :2])

ax1.plot(x, np.sin(x))
ax1.plot(x, np.cos(x))
ax3.plot(np.cos(x), x)
ax3.plot(np.sin(x), x)
ax2.plot(x,x**2)

plt.show()
```
